#### OOP

```python
class Robot:
    def __init__(self, id, name, year, color="Black"):
        self.id = id
        self.name = name
        self.year = year
        self.color = color
        
    def getId(self):
        return self.id
        
    def getName(self):
        return self.name
    
    def getYear(self):
        return self.year
    
    def setName(self, name):
        self.name = name
    
    def getProperties(self):
        return self.id, self.name, self.year, self.color

# add "_" for large numbers to increase readability
robot_0 = Robot(10_120_000, "Spot", 1995, "Green")
robot_1 = Robot(1_120_000, "Tops", 1987, "Red")
print(f"Robot 0 id: {robot_0.getId()}")

# Tuple unpacking manipulation with "*" and "_"
_, *desc, color = robot_0.getProperties()
print(desc)

# Lambda function
newer_robot = lambda x, y: x if (x.getYear() > y.getYear()) else y
print(newer_robot(robot_0, robot_1).getName())
```

#### Catch an error

```python
from getpass import getpass # hide user input

try: 
    val = getpass("Give me a positive number! ")
    num = int(val)
    
    assert num > 0
    
    print("Oh you are good!")
    
except ValueError:
    print("I. SAID. A. NUMBER.")
    
except AssertionError:
    print("That was a negative number my friend!")
    
finally:
    print("That was fun! I dare you to play again")
```

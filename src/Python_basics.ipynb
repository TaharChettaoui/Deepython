#### OOP

```python
class Robot:
    robots = []
    
    def __init__(self, id, name, year, color="Black"):
        self.id = id
        self._name = name
        self.year = year
        self.color = color
        Robot.robots.append(self)
        
    def _getName(self):
        return self._name
    
    def _setName(self, name):
        self._name = name
        
    def getYear(self):
        return self.year
    
    def constructedRobots(cls):
        return cls.robots
    
    
    name = property(_getName, _setName)

Robot.constructedRobots = classmethod(Robot.constructedRobots)
```

```python
# add "_" for large numbers to increase readability
robot_0 = Robot(10_120_000, "Spot", 1995, "Green")
robot_1 = Robot(1_120_000, "Tops", 1987, "Red")
robot_2 = Robot(1_130_000, "Müller", 1996, "Yellow")
```

```python
# Tuple unpacking manipulation with "*" and "_"
_, *desc, color = robot_0.__dict__

print("Some of Robot 0 properties: " + str(desc))
```

```python
# Lambda function
newer_robot = lambda x, y: x if (x.getYear() > y.getYear()) else y

print("Newest robot is : " + newer_robot(robot_0, robot_1).name)
```

```python
# List comprehension
print("These robots were built after the year 1990: ")
[rbt.name for rbt in Robot.constructedRobots() if rbt.getYear() > 1990]
```

#### Catch an error
User interactiv "game" in order to catch errors

```python
from getpass import getpass # hide user input

try: 
    val = getpass("Give me a positive number! ")
    num = int(val)
    
    assert num > 0
    
    print("Oh you are good!")
    
except ValueError:
    print("I. SAID. A. NUMBER.")
    
except AssertionError:
    print("That was a negative number my friend!")
    
finally:
    print("That was fun! Lets play again")
```

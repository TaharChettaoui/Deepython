- ___ slots ___ : Python uses the dictionaries to manage the instance attributes. It allows you to add more attributes dynamically. but has a certain memory overhead. "slots" optimizes the memory if the class has many objects. (around 20% performance improvement). SLOTS DO NOT WORK WITH MULTIPLE INHERITANCE! (logical)
- __Descriptors__ : are Python objects that implement a method of the descriptor protocol (get, set, delete, set_name). If your descriptor implements just get(), then it’s a non-data descriptor. If it implements set() or delete(), then it’s said to be a data descriptor.

```python
from dataclasses import dataclass

@dataclass #(slots=True) for python > 3.10
class Person:
    name: str
    surname: str
    adress: str
        
class Robot:
    _robots = []
    __slots__ = "id", "_name", "owner","year", "color"
    
    def __init__(self, id, name, owner: Person, year, color="Black"):
        self.id = id
        self.owner = owner
        self._name = name
        self.year = year
        self.color = color
        Robot._robots.append(self)
        
    def _get_name(self):
        return self._name
    
    def _set_name(self, name):
        self._name = name
    
    def constructed_robots(cls):
        return cls._robots
    
    
    name = property(_get_name, _set_name)

Robot.constructed_robots = classmethod(Robot.constructed_robots)
```

```python
person_0 = Person("Thomas", "Müller", "76131 Karlsruhe")

# add "_" for large numbers to increase readability
robot_0 = Robot(20_120_000, "Spot", person_0, 1995, "Green")
robot_1 = Robot(10_100_000, "Tops", person_0, 1987, "Red")
robot_2 = Robot(20_130_000, "Müller", person_0, 1996, "Yellow")
```

```python
# Tuple unpacking manipulation with "*" and "_"
_, *desc, color = robot_0.__slots__

print("Some of Robot 0 properties: " + str(desc))
```

```python
# Lambda function
newer_robot = lambda x, y: x if (x.year > y.year) else y

print("Newest robot is : " + newer_robot(robot_0, robot_1).name)
```

```python
# List comprehension
print("These robots were built after the year 1990: ")
[rbt.name for rbt in Robot.constructed_robots() if rbt.year > 1990]
```

#### Catch an error
User interactiv "game" in order to catch errors

```python
from getpass import getpass # hide user input

try: 
    val = getpass("Give me a positive number! ")
    num = int(val)
    
    assert num > 0
    
    print("Oh you are good!")
    
except ValueError:
    print("I. SAID. A. NUMBER.")
    
except AssertionError:
    print("That was a negative number my friend!")
    
finally:
    print("That was fun! Lets play again")
```

## 7. Filtering

```python
# example 1
filtered_age = (df['Age'] == "18-24 years old") & (df['Employment'] == "Employed, full-time") # or |, and &, negate ~
df.loc[filtered_age, 'ConvertedCompYearly'] # by adding ~

# example 2
countries = ['France', 'Germany']
high_salary = (df['Age'] == "18-24 years old") & (df['ConvertedCompYearly'] > 0) & (df['Country'].isin(countries))
df.loc[high_salary].head(2)
```

## 9. Updating Rows and Columns

```python
# .str because it's a Series
df['MainBranch'] = df['MainBranch'].str.lower()

# apply
def update_q120(q120_status):
    return q120_status.upper()

df['Q120'].apply(update_q120)

# alternative with lambda function
df['Q120'].apply(lambda x: x.lower())
```

```python
# map
df[['Q120', 'MainBranch']].map(len)
```

## 10. Add/Remove rows and columns

```python
### Rows
# ADD
new_row = {'Q120': 'I do not agree'}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

# DEL with index
# num_rows = df.shape[0] - 1
# df = df.drop(index=num_rows)

# DEL with filter
filt = df['Q120'] == "I do not agree"
df = df.drop(index=df[filt].index) # include all occurences
```

```python
### Column
# ADD
df['AgeRemote'] = df['Age'] + ' ' + ['RemoteWork']

# DEL
# df = df.drop(columns=['Age', 'RemoteWork'])

# SPLIT
#df[['Age', 'RemoteWork']] = df['AgeRemote'].str.split(' ', expand=True)
```

## 11. Data cleaning 

```python
# clean NaN values 
df.dropna(axis='index', how='any') # default axis is 'index', drop rows with any missing values

# df.dropna(axis='columns', how='all')
df.dropna(axis='index', how='any', subset=['Age', 'Employment'])

# Deal with missing values 
df.replace('Missing', np.nan, inplace=True)
df.isna()
df.fillna('MISSING').head(2)
```

## 12. Dates and Time Series Data

https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

```python
# Convert input csv dates to the right dates format 
database_eth_path = '../dataset/ETH_1h.csv'
df_eth = pd.read_csv(database_eth_path, parse_dates=['Date'], date_format='%Y-%m-%d %I-%p')

# Alternative solution
# df_eth['Date'] = pd.to_datetime(df_eth['Date'], format=%Y-%m-%d)

# Add column with day of the week according to the date
df_eth['DayOfWeek'] = df_eth['Date'].dt.day_name()
# df_eth.loc[0, 'order_date'].day_name()

# min/max date
df_eth['Date'].min()
df_eth['Date'].max()
df_eth['Date'].max() - df_eth['Date'].min() 

filt = (df_eth['Date'] >= pd.to_datetime('2019-01-01')) & (df_eth['Date'] <= pd.to_datetime('2020-01-01'))

df_eth.loc[filt]['Close'].mean()
```

```python
df_eth.set_index('Date', inplace=True)
highs = df_eth['High'].resample('D').max()
highs
```

```python
%matplotlib inline
highs.plot()
```

```python
df_eth.resample('W').agg({'Close': 'mean', 'High': 'max', 'Low': 'min', 'Volume':'sum'})
```
